```python


print("Dataset shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

df.describe(include='all')
```

```python


df.drop_duplicates(inplace=True)


df.isnull().sum()
```

```python

# Ensure 'Sales' is numeric before plotting
df['Sales'] = df['Sales'].replace('[\$,]', '', regex=True).astype(float)

df.groupby("Region")["Sales"].sum().plot(kind='bar', title="Total Sales by Region")
plt.ylabel("Total Sales")
plt.show()

df['Region'].value_counts().plot(kind='bar', title="Number of Orders by Region")
plt.ylabel("Number of Orders")
plt.show()

df.groupby("Region")["Sales"].median().plot(kind='bar', title="Median Sales by Region")
plt.ylabel("Median Sales")
plt.show()
```

```python
if 'Sub-Category' in df.columns and 'Profit' in df.columns:
    plt.figure(figsize=(10, 5))
    df.groupby('Sub-Category')['Profit'].sum().sort_values().plot(kind='bar', color='teal')
    plt.title("Total Profit by Sub-Category")
    plt.xlabel("Sub-Category")
    plt.ylabel("Total Profit")
    plt.tight_layout()
    plt.show()
else:
    print("Required columns 'Sub-Category' and/or 'Profit' not found in dataset.")
```

```python

df['Sales'] = df['Sales'].replace('[\$,]', '', regex=True).astype(float)
df['Profit'] = df['Profit'].replace('[\$,]', '', regex=True).astype(float)

# Plot 1: Histogram of Sales
plt.figure(figsize=(8, 4))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Distribution of Sales")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()

# Plot 2: Boxplot of Sales by Category (change 'Category' if you have 'Gender' instead)
plt.figure(figsize=(6, 4))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

# Plot 3: Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python


if 'Payment' in df.columns:
    df.groupby("Payment")["Sales"].mean().plot(kind='bar', title="Avg Sales by Payment Method", figsize=(6, 4))
    plt.ylabel("Average Sales")
    plt.xlabel("Payment Method")
    plt.show()
else:
    print("Column 'Payment' not found in dataset.")


date_column = 'Date' if 'Date' in df.columns else 'Order Date'

if date_column in df.columns:
    df[date_column] = pd.to_datetime(df[date_column])
    daily_sales = df.groupby(date_column)["Sales"].sum()
    daily_sales.plot(title="Sales Trend Over Time", figsize=(10, 4))
    plt.ylabel("Total Sales")
    plt.xlabel("Date")
    plt.show()
else:
    print("No suitable date column found.")
```
